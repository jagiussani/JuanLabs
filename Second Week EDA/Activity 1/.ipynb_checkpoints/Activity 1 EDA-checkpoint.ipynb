```python
import numpy as np
import pandas as pd
file1_df = pd.read_csv('file1.csv')
file2_df = pd.read_csv('file2.csv')
file3_df = pd.read_csv('file3.csv')
```

```python
############# Uniforming the columns from the dataframes #################
#Lower case all the columns

def lower_case_column_names(file1_df):
    file1_df.columns=[i.lower() for i in file1_df.columns]
    return file1_df
```

```python
file1_df=lower_case_column_names(file1_df)
file1_df.head()
```

```python
file2_df=lower_case_column_names(file2_df)
file2_df.head()
```

```python
file3_df=lower_case_column_names(file3_df)
file3_df.head()
```

```python
# renaming columns File 1
def rename_columns(file1_df):
    file1_df.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value',
                          'monthly premium auto':'monthly_premium_auto',"number of open complaints":"number_of_open_complaints", 
                          "policy type":"policy_type","vehicle class": "vehicle_class",
                          "total claim amount":"total_claim_amount"}, inplace=True )
    return file1_df
file1_df=rename_columns(file1_df)

file1_df.head()

# Try to rewrite the function to give a local variable a different name as the global one

```

```python
# renaming columns File 2
file2_df=rename_columns(file2_df)

file2_df.head()
```

```python
# renaming columns File 3
file3_df=rename_columns(file3_df)

file3_df.head()
```

```python
#integrating all 3 files - concatanating 
file_t=pd.concat([file1_df,file2_df,file3_df], axis=0)
file_t
```

```python
file_t.info()
```

```python
########### Dropping dupplicates #############
file_t = file_t.drop_duplicates()
file_t
# 2.939 duplicates got dropped
```

```python
# After concatanating - reset data frame
file_t=file_t.reset_index()
file_t
```

```python
# Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data
def drop_columns(file_t) :
    file_t.drop(columns=["customer","index"], inplace=True)
    return file_t
file_t= drop_columns(file_t)
file_t
```

```python
# Working with data types – Check the data types of all the columns and fix the incorrect ones 
# (for ex. customer lifetime value and number of open complaints ). 
# Hint: remove the percentage from the customer lifetime value and truncate it to an integer value.

file_t['customer_lifetime_value'] = file_t['customer_lifetime_value'].str.strip('%')       # removing % whiting the customer_lifetime_value
file_t
```

```python
#clean the number of open complaints and extract the middle number which is changing between records.
# hint: combine lambda and if functions example lambda x: x.lower() if type(x) == str else x

file_t['number_of_open_complaints'] = file_t['number_of_open_complaints'].apply(lambda x: str(x.split('/')[1]) if '/' in str(x) else x)
file_t.head()
```

```python
##### replace nan with 0
file_t['number_of_open_complaints'] = file_t['number_of_open_complaints'].fillna(value=0)
```

```python
file_t['number_of_open_complaints'] = file_t['number_of_open_complaints'].astype(float)
file_t['number_of_open_complaints'] = file_t['number_of_open_complaints'].astype(int)
file_t
```
