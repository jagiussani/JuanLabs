#### Uniforming the columns from the dataframes

```python
#Lower case all the columns
def lower_case_column_names(file1_df):
    file1_df.columns=[i.lower() for i in file1_df.columns]
    return file1_df
```

```python
file1_df=lower_case_column_names(file1_df) # step 1 only for file 1
file1_df.head()
```

```python
file2_df=lower_case_column_names(file2_df) # step 2
file2_df.head()
```

```python
file3_df=lower_case_column_names(file3_df) # step 3
file3_df.head()
```

```python
# renaming columns File 1
def rename_columns(file1_df):
    file1_df.rename(columns={'st':'state','customer lifetime value':'customer_lifetime_value',
                          'monthly premium auto':'monthly_premium_auto',"number of open complaints":"number_of_open_complaints", 
                          "policy type":"policy_type","vehicle class": "vehicle_class",
                          "total claim amount":"total_claim_amount"}, inplace=True )
    return file1_df
file1_df=rename_columns(file1_df)

file1_df.head()

# Try to rewrite the function to give a local variable a different name as the global one

```

```python
# renaming columns File 2
file2_df=rename_columns(file2_df)

file2_df.head()
```

```python
# renaming columns File 3
file3_df=rename_columns(file3_df)

file3_df.head()
```

##### Let us chnage the data type of Columns in the different file in order to have an ordered concatanation

```python
# Convert columns of file 2 into float, as in file 1
file2_df[['income', 'monthly_premium_auto', 'total_claim_amount']] = file2_df[['income', 'monthly_premium_auto', 'total_claim_amount']].astype(float)
file2_df.info()
```

```python
# Convert columns of file 3 into float, as in file 1 and 2
file3_df[['income', 'monthly_premium_auto']] = file3_df[['income', 'monthly_premium_auto']].astype(float)
file3_df['number_of_open_complaints'] = file3_df['number_of_open_complaints'].astype(float)
file3_df.info()
```

```python
# Last but not least: lets change two columns of file 1 and 2, so that it mathces file 3
file3_df[['customer_lifetime_value','number_of_open_complaints']] = file3_df[['customer_lifetime_value','number_of_open_complaints']].astype(str)
```

#### Concat the files

```python
#integrating all 3 files - concatanating 
file_t=pd.concat([file1_df,file2_df,file3_df], axis=0)
file_t
```

```python
file_t.info()
```

```python
nancount_filet = file_t.isna().sum()
print(nancount_filet)
```

```python
########### Dropping dupplicates #############
file_t = file_t.drop_duplicates()
file_t
# 2.939 duplicates got dropped
```

```python
# After concatanating - reset data frame
file_t=file_t.reset_index()
file_t
```

```python
# Deleting and rearranging columns – delete the column customer as it is only a unique identifier for each row of data
def drop_columns(file_t) :
    file_t.drop(columns=["customer","index"], inplace=True)
    return file_t
file_t= drop_columns(file_t)
file_t
```

### Replacing null values with the means of the column

```python
file_t.head()
```

```python
# Replacing null values – Replace missing values with means of the column (for numerical columns). 
# Pay attention that the Income feature for instance has 0s which is equivalent to null values.
#(We assume here that there is no such income with 0 as it refers to missing values) 
# Hint: numpy.nan is considered of float64 data type.

#Numerical columns: 
## customer_lifetime_value, income, monthly_premium_auto, number_of_open_complaints, total_claim_amount

median_values = list(np.mean(file_t[['customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']]))
print(median_values)   
```

```python
median_dict = {'customer_lifetime_value': median_values[0],
               'income': median_values[1],
               'monthly_premium_auto': median_values[2],
               'number_of_open_complaints': median_values[3],
               'total_claim_amount': median_values[4]}
```

```python
file_t[['customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']] = file_t[['customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']].fillna(value=median_dict)
```
